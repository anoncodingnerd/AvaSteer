```python
import numpy as np
import pandas as pd
```

```python
mathv = pd.read_json("Data/Questions/MathV.jsonl", lines=True)
mathvmini = pd.read_json("Data/Questions/MathVMini.jsonl", lines=True)

full_ids = mathv["id"].astype(str)
mini_ids = set(mathvmini["id"].astype(str))
train_mask = ~full_ids.isin(mini_ids)
train_idx = np.flatnonzero(train_mask)
print(f"train size: {train_idx.size} / {len(mathv)}")
```

```python
from pathlib import Path
import re

embed_dir = Path("Data/Representation/MathV/GLM-4.1V-9B-Thinking-no-image")
out_dir = Path("Data/Representation/MathV_split/GLM-4.1V-9B-Thinking-no-image")
out_dir.mkdir(parents=True, exist_ok=True)

def _sort_key(p):
    m = re.search(r"(\d+)", p.stem)
    return int(m.group(1)) if m else p.name

embed_paths = sorted(embed_dir.glob("embeds_*.npy"), key=_sort_key)

train_embeds = {}
for p in embed_paths:
    emb = np.load(p)
    train_embeds[p.name] = emb[train_idx]
    np.save(out_dir / p.name, train_embeds[p.name])

print({k: v.shape for k, v in train_embeds.items()})
print(f"saved to: {out_dir}")
```
